```python
# Install required libraries (if not already installed)
!pip install pandas numpy scikit-learn openpyxl

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# -----------------------------
# 1. Create Historical Dataset
# -----------------------------
data = {
    "Year": [2019, 2020, 2021, 2022, 2023, 2024],
    "April_Avg": [32.0, 32.5, 33.0, 33.2, 33.5, 34.0],
    "May_Avg": [34.5, 35.0, 35.5, 36.0, 36.5, 37.0],
    "June_Avg": [32.8, 32.4, 33.1, 32.9, 33.4, 33.2],
    "Peak_Temp": [36.5, 38.0, 39.5, 40.0, 41.0, 43.0]
}

df = pd.DataFrame(data)

print("Historical Data:")
print(df)

# -----------------------------
# 2. Train Model for Prediction
# -----------------------------
X = df[["Year"]]

predictions = {}

months = ["April_Avg", "May_Avg", "June_Avg", "Peak_Temp"]

for month in months:
    y = df[month]

    model = LinearRegression()
    model.fit(X, y)

    pred_2026 = model.predict([[2026]])[0]
    predictions[month] = round(pred_2026, 2)

# -----------------------------
# 3. Create Prediction DataFrame
# -----------------------------
pred_df = pd.DataFrame({
    "Year": [2026],
    "April_Avg": [predictions["April_Avg"]],
    "May_Avg": [predictions["May_Avg"]],
    "June_Avg": [predictions["June_Avg"]],
    "Peak_Temp": [predictions["Peak_Temp"]]
})

print("\nPredicted Data (2026):")
print(pred_df)

# -----------------------------
# 4. Save to Excel File
# -----------------------------
with pd.ExcelWriter("Kolkata_Summer_Temp_Data.xlsx") as writer:
    df.to_excel(writer, sheet_name="Historical_Data", index=False)
    pred_df.to_excel(writer, sheet_name="Prediction_2026", index=False)

print("\nExcel file saved successfully!")

# -----------------------------
# 5. (Optional) Plot Graph
# -----------------------------
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(df["Year"], df["Peak_Temp"], marker='o', label="Historical Peak Temp")
plt.scatter(2026, predictions["Peak_Temp"], color='red', label="Predicted 2026")

plt.title("Kolkata Summer Temperature Trend")
plt.xlabel("Year")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.grid()

plt.show()
```
