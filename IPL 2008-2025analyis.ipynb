```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/ipl-indian-premier-league-2008-to-2025-dataset.zip')
df.head()


df['Winner'].value_counts()
df['Player of the Tournament'].value_counts().head(10)
df['Winner'].value_counts()


plt.figure(figsize=(14, 8))
sns.barplot(x='Season', y='Most Runs', data=df, palette='viridis')
plt.title('Most Runs (Orange Cap) by Season (2008-2025)')
plt.xlabel('Season')
plt.ylabel('Total Runs')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate with player names
for index, row in df.iterrows():
    plt.text(index, row['Most Runs'] + 10, row['Most Runs Player'], color='black', ha='center', va='bottom', fontsize=9, rotation=90)

plt.tight_layout()
plt.show()
plt.figure(figsize=(14, 8))
sns.barplot(x='Season', y='Most Wickets', data=df, palette='magma')
plt.title('Most Wickets (Purple Cap) by Season (2008-2025)')
plt.xlabel('Season')
plt.ylabel('Total Wickets')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate with player names
for index, row in df.iterrows():
    plt.text(index, row['Most Wickets'] + 0.5, row['Most Wickets Player'], color='black', ha='center', va='bottom', fontsize=9, rotation=90)

plt.tight_layout()
plt.show()



plt.figure(figsize=(10, 8))
df['Winner'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, cmap='Pastel1')
plt.title('Proportion of IPL Wins by Team (2008-2025)')
plt.ylabel('') # Hide the default 'Winner' label
plt.tight_layout()
plt.show()


common_players_mr_pot = df[df['Most Runs Player'] == df['Player of the Tournament']]
print("Players who won both 'Most Runs Player' and 'Player of the Tournament' in the same season:")
if not common_players_mr_pot.empty:
    display(common_players_mr_pot[['Season', 'Most Runs Player']])
else:
    print("No players won both in the same season.")

common_players_mw_pot = df[df['Most Wickets Player'] == df['Player of the Tournament']]
print("Players who won both 'Most Wickets Player' and 'Player of the Tournament' in the same season:")
if not common_players_mw_pot.empty:
    display(common_players_mw_pot[['Season', 'Most Wickets Player']])
else:
    print("No players won both in the same season.")


highest_runs_row = df.loc[df['Most Runs'].idxmax()]
print(f"Highest 'Most Runs' in a single season: {highest_runs_row['Most Runs']} by {highest_runs_row['Most Runs Player']} in {highest_runs_row['Season']}")

lowest_runs_row = df.loc[df['Most Runs'].idxmin()]
print(f"Lowest 'Most Runs' in a single season: {lowest_runs_row['Most Runs']} by {lowest_runs_row['Most Runs Player']} in {lowest_runs_row['Season']}")

highest_wickets_row = df.loc[df['Most Wickets'].idxmax()]
print(f"Highest 'Most Wickets' in a single season: {highest_wickets_row['Most Wickets']} by {highest_wickets_row['Most Wickets Player']} in {highest_wickets_row['Season']}")

unique_captains_count = df['Captain'].nunique()
print(f"Total number of unique captains over all seasons: {unique_captains_count}")

lowest_wickets_row = df.loc[df['Most Wickets'].idxmin()]
print(f"Lowest 'Most Wickets' in a single season: {lowest_wickets_row['Most Wickets']} by {lowest_wickets_row['Most Wickets Player']} in {lowest_wickets_row['Season']}")

plt.figure(figsize=(10, 6))
sns.histplot(df['Most Runs'], kde=True, bins=5, color='orange')
plt.title('Distribution of \'Most Runs\' (Orange Cap) Scores (2008-2025)')
plt.xlabel('Total Runs')
plt.ylabel('Number of Seasons')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df['Most Wickets'], kde=True, bins=5, color='purple')
plt.title('Distribution of \'Most Wickets\' (Purple Cap) Counts (2008-2025)')
plt.xlabel('Total Wickets')
plt.ylabel('Number of Seasons')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(y='Final Man of the Match', data=df, order=df['Final Man of the Match'].value_counts().index, palette='rocket', legend=False)
plt.title('Players with Most \'Final Man of the Match\' Awards (2008-2025)')
plt.xlabel('Number of Awards')
plt.ylabel('Player')
plt.tight_layout()
plt.show()

winning_captains = df['Captain'][df['Winner'] == df['Winner']]

plt.figure(figsize=(10, 6))
sns.countplot(y=winning_captains, order=winning_captains.value_counts().index, palette='mako', legend=False)
plt.title('Captains with Most IPL Wins (2008-2025)')
plt.xlabel('Number of Wins')
plt.ylabel('Captain')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 7))
sns.barplot(x='Season', y='Winner', data=df, palette='Spectral')
plt.title('IPL Winners by Season (2008-2025)')
plt.xlabel('Season')
plt.ylabel('Winning Team')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()


# Set a style for the plots
sns.set_style('whitegrid')

# 1. Top players with 'Player of the Tournament' awards
plt.figure(figsize=(10, 6))
sns.countplot(y='Player of the Tournament', data=df, order=df['Player of the Tournament'].value_counts().index, palette='viridis')
plt.title('Players with Most \'Player of the Tournament\' Awards (2008-2025)')
plt.xlabel('Number of Awards')
plt.ylabel('Player')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(y='Most Runs Player', data=df, order=df['Most Runs Player'].value_counts().index, palette='magma')
plt.title('Players with Most \'Most Runs Player\' Awards (Orange Cap) (2008-2025)')
plt.xlabel('Number of Awards')
plt.ylabel('Player')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(y='Most Wickets Player', data=df, order=df['Most Wickets Player'].value_counts().index, palette='cubehelix')
plt.title('Players with Most \'Most Wickets Player\' Awards (Purple Cap) (2008-2025)')
plt.xlabel('Number of Awards')
plt.ylabel('Player')
plt.show()

plt.figure(figsize=(12, 7))
sns.lineplot(x='Season', y='Most Runs', data=df, marker='o', color='orange')
plt.title('Most Runs (Orange Cap Holder) Trend Over IPL Seasons (2008-2025)')
plt.xlabel('Season')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 7))
sns.lineplot(x='Season', y='Most Wickets', data=df, marker='o', color='purple')
plt.title('Most Wickets (Purple Cap Holder) Trend Over IPL Seasons (2008-2025)')
plt.xlabel('Season')
plt.ylabel('Total Wickets')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
